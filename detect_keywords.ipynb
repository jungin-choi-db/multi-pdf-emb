```python
import torch
from torchvision import models, transforms
from PIL import Image
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from datetime import datetime

# 이미지 분류 모델 로드
model = models.densenet121(pretrained=True)
model.eval()

# 클래스 이름 로드
with open("imagenet_classes.txt") as f:
    labels = [line.strip() for line in f.readlines()]

# 클래스 이름 길이 출력 (디버깅용)
print(f"Number of labels: {len(labels)}")

# 이미지 전처리
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def classify_image(image_path):
    # 이미지 열기
    input_image = Image.open(image_path).convert('RGB')

    # 이미지 전처리
    input_tensor = preprocess(input_image)
    input_batch = input_tensor.unsqueeze(0)

    # GPU 사용 가능 시 GPU로 이동
    if torch.cuda.is_available():
        input_batch = input_batch.to('cuda')
        model.to('cuda')

    # 모델 추론
    with torch.no_grad():
        output = model(input_batch)

    # 결과 해석
    probabilities = torch.nn.functional.softmax(output[0], dim=0)
    top5_prob, top5_catid = torch.topk(probabilities, 5)

    # 상위 5개 클래스와 확률 출력 (디버깅용)
    print(f"Top 5 probabilities: {top5_prob}")
    print(f"Top 5 category IDs: {top5_catid}")

    # 상위 5개 클래스와 확률 출력
    top5_labels = []
    for i in range(top5_prob.size(0)):
        cat_id = top5_catid[i].item()
        # cat_id가 labels 리스트의 범위 내에 있는지 확인
        if cat_id < len(labels):
            top5_labels.append(labels[cat_id])
        else:
            print(f"Warning: Category ID {cat_id} is out of range for labels list.")
    return top5_labels

# 이미지 분류 예시
image_path = '/content/sample.jpg'  # 이미지를 업로드한 경로로 수정
detected_labels = classify_image(image_path)
print("Detected Labels:", detected_labels)
```

```python
# GPT 모델과 토크나이저 불러오기
model_name = 'gpt2-medium'
gpt_model = GPT2LMHeadModel.from_pretrained(model_name)
gpt_tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# 상황 인식 결과를 이용해 일기 생성 함수
def generate_diary_entry(labels):
    # 오늘 날짜 가져오기
    today = datetime.now().strftime("%Y-%m-%d")
    # 레이블을 하나의 문자열로 결합
    prompt = f"Date: {today}\nToday's event is about " + ", ".join(labels) + ".\n "

    # 텍스트 생성
    inputs = gpt_tokenizer.encode(prompt, return_tensors='pt')
    outputs = gpt_model.generate(inputs, max_length=300, num_return_sequences=1, no_repeat_ngram_size=2, top_p=0.95, temperature=0.85)

    # 생성된 텍스트 디코딩
    generated_text = gpt_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# 키워드를 이용해 일기 작성 예시
diary_entry = generate_diary_entry(detected_labels)
print(" **** Diary ****\n", diary_entry)
```
